### Creating Data Set

```python
import cv2
import os

# Define the directory to store images
output_folder = r"C:\Users\admin\Desktop\Akh all doc\M.Sc. Sem IV\Advance Topics in AI\Practical\Pract 1\CNN_data\Anant"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Initialize the webcam
cap = cv2.VideoCapture(0)
# Load OpenCV's pre-trained face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

count = 0  # Counter for saved images
max_images = 100  # Number of images to collect

print("[INFO] Collecting face data. Press 'q' to exit.")

while count < max_images:
    ret, frame = cap.read()
    if not ret:
        print("[ERROR] Failed to capture image.")
        break
    # Convert to grayscale for better face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(50, 50))

    for (x, y, w, h) in faces:
        face = frame[y:y+h, x:x+w]  # Extract the face
        face_resized = cv2.resize(face, (100, 100))  # Resize to a fixed size

        # Save face image
        img_name = os.path.join(output_folder, f"face_{count}.jpg")
        cv2.imwrite(img_name, face_resized)
        count += 1

        # Draw a rectangle around detected face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        print(f"[INFO] Image {count}/{max_images} saved: {img_name}")

    # Show live webcam feed
    cv2.imshow("Face Collection", frame)

    # Press 'q' to exit early
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

print("[INFO] Face data collection complete.")

# Release resources
cap.release()
cv2.destroyAllWindows()
```

### Training and Validation Split

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Conv2D, BatchNormalization, Activation, MaxPooling2D, Add, Flatten, Dense, Input, GlobalAveragePooling2D
)
import os

# Define paths
dataset_path = r"C:\Users\admin\Desktop\Akh all doc\M.Sc. Sem IV\Advance Topics in AI\CH III\CNN_data"  # Path to dataset
image_size = (224, 224)  # Standard ResNet size
batch_size = 32
num_classes = len(os.listdir(dataset_path))  # Count number of folders (students)

# Data Augmentation & Preprocessing
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # 20% for validation
)

train_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=image_size,
    batch_size=batch_size,
    class_mode="categorical",
    subset="training"
)

val_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=image_size,
    batch_size=batch_size,
    class_mode="categorical",
    subset="validation"
)
```

### ResNET50 Architecture From Scratch

```python
from tensorflow.keras.layers import Conv2D, BatchNormalization, Activation, Add

def residual_block(x, filters, downsample=False):
    """
    Creates a residual block with an optional downsampling shortcut. 
    Args:
    - x: Input tensor
    - filters: Number of filters for Conv layers
    - downsample: If True, apply a 1x1 Conv layer to match dimensions
    Returns:
    - Output tensor after residual block
    """
    shortcut = x  # Store input for residual connection
    # First convolution
    x = Conv2D(filters, kernel_size=(3, 3), padding="same", strides=(2 if downsample else 1))(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    # Second convolution
    x = Conv2D(filters, kernel_size=(3, 3), padding="same")(x)
    x = BatchNormalization()(x)

    # Adjust shortcut if downsampling or filter size changes
    if downsample or shortcut.shape[-1] != filters:
        shortcut = Conv2D(filters, kernel_size=(1, 1), strides=(2 if downsample else 1), padding="same")(shortcut)
        shortcut = BatchNormalization()(shortcut)

    # Add shortcut and apply activation
    x = Add()([x, shortcut])
    x = Activation("relu")(x)
    return x

def build_resnet50(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    # Initial Conv Layer
    x = Conv2D(64, kernel_size=(7, 7), strides=(2, 2), padding="same")(inputs)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D(pool_size=(3, 3), strides=(2, 2), padding="same")(x)

    # Residual Blocks
    x = residual_block(x, 64)
    x = residual_block(x, 64)
    
    x = residual_block(x, 128, downsample=True)  # Fix: Use downsample when changing filters
    x = residual_block(x, 128)
    
    x = residual_block(x, 256, downsample=True)  # Fix
    x = residual_block(x, 256)
    
    x = residual_block(x, 512, downsample=True)  # Fix
    x = residual_block(x, 512)

    # Global Pooling and Dense Layer
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation="relu")(x)
    outputs = Dense(num_classes, activation="softmax")(x)

    # Create Model
    model = Model(inputs, outputs, name="ResNet50_Scratch")
    return model
```

```python
model = build_resnet50(input_shape=(224, 224, 3), num_classes=num_classes)
model.summary()
# Compile model
model.compile(
    loss="categorical_crossentropy",
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    metrics=["accuracy"]
)
# Train the Model
epochs = 50 
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=epochs,
    verbose=1
)
# Save Model
model.save("resnet50_scratch_face_classifier.h5")
print("[INFO] Training complete. Model saved as resnet50_scratch_face_classifier.h5")
```

```python
loss, accuracy = model.evaluate(val_generator)
print(f"Validation Accuracy: {accuracy*100:.2f}%")
```

### Sample Image Recognition By Model

```python
import numpy as np
from tensorflow.keras.preprocessing import image
# Load Model
model = tf.keras.models.load_model("resnet50_scratch_face_classifier.h5")

# Load and preprocess image
img_path = r'C:\Users\admin\Desktop\Akh all doc\M.Sc. Sem IV\Advance Topics in AI\CH III\CNN_data\Akhilesh\face_96.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

# Make Prediction
prediction = model.predict(img_array)
predicted_class_index = np.argmax(prediction)
# Reverse dictionary mapping
class_labels = {v: k for k, v in train_generator.class_indices.items()}
print(f"Predicted Class: {class_labels[predicted_class_index]}\n")

img = cv.imread(img_path, 1)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
true_labels = []
predicted_labels = []
for i in range(len(val_generator)):
    x_batch, y_batch = val_generator[i]  # Get images and labels
    predictions = model.predict(x_batch)  # Predict on batch
    true_labels.extend(np.argmax(y_batch, axis=1))  # Convert one-hot to label index
    predicted_labels.extend(np.argmax(predictions, axis=1))  # Get predicted label index

true_labels = np.array(true_labels)
predicted_labels = np.array(predicted_labels)
# Get class names from generator
class_labels = list(train_generator.class_indices.keys())
```

```python
true_labels_train = []
predicted_labels_train = []
for i in range(len(train_generator)):
    x_batch, y_batch = train_generator[i]  # Get images and labels
    predictions = model.predict(x_batch)  # Predict on batch
    true_labels_train.extend(np.argmax(y_batch, axis=1))  # Convert one-hot to label index
    predicted_labels_train.extend(np.argmax(predictions, axis=1))  # Get predicted label index

true_labels_train = np.array(true_labels_train)
predicted_labels_train = np.array(predicted_labels_train)
```
